```python
import pandas as pd
import matplotlib.pyplot as plt
from plot import plot ,plot_bool
data = pd.read_csv("data.csv")
```

## Meiste Bevölkerung 2010 vs 2020

```python
plot(data["city"] ,data["population_2010"] ,"2010" ,8 ,True)
plot(data["city"] ,data["population_2020"] ,"2020" ,8 ,True)
```

## Wenigste Bevölkerung 2010 vs 2020

```python
plot(data["city"] ,data["population_2010"] ,"2010" ,8 ,False)
plot(data["city"] ,data["population_2020"] ,"2020" ,8 ,False)
```

## Fläche

```python
plot(data["city"] ,data["land_area_sqkm"] ,"Flächengrößte Städte in km2" ,8 ,True)
plot(data["city"] ,data["land_area_sqkm"] ,"Flächenkleinste Städte in km2" ,8 ,False)
```

## Populationsdichte

```python
plot(data["city"] ,data["pop_density_sqkm"] ,"Meiste Population pro km2" ,8 ,True)
plot(data["city"] ,data["pop_density_sqkm"] ,"Wenigste Population pro km2" ,8 ,False)
```

## Größte Stadt im bezüglichen Staat

```python
plot_bool(["Größte Stadt" ,"Nicht Größte Stadt"] ,data["largest_city_in_state"] ,"Größte Stadt im Staat")
```

## Staatshauptstadt in bezüglichen Staat

```python
plot_bool(["Staatshauptstadt" ,"Nicht Staatshauptstadt"] ,data["state_capital"] ,"Staatshauptstadt in bezüglichen Staat")
```
